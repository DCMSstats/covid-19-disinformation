# Init

This needs to be run the first time to initialise the tables. 

This is due to the progress appending data onto existing data and that the SQL query fails due to the table having no schema. 

```python
import pandas as pd
import pandas_gbq
import praw
import numpy as np
from prawcore import PrawcoreException
import datetime
import json
import disinfo as di
```

```python
with open("gcp_config.json") as f:
    config = json.load(f)

project_id = config["project_id"]
table_sub  = config["table_sub"]
table_com  = config["table_com"]

comments_number = 10
topics_list = config["topics_list"]

client_id = config["client_id"]
client_secret = config["client_secret"]
user_agent =  config["user_agent"]

# init the praw instance
reddit = praw.Reddit(client_id = client_id, client_secret = client_secret,  user_agent = user_agent)
```

```python
# Collect the reddit data 

current_data = di.get_reddit(topics_list, comments_number, reddit_inst=reddit)
current_data["created"] = current_data["created"].apply(di.convert_date)
current_data["backup_date"] = datetime.date.today()
current_data["backup_time"] = datetime.datetime.today().time()
print("Subreddit data collected")
```

```python
# Write Reddit data to bigquery
pandas_gbq.to_gbq(current_data, table_sub, project_id=project_id, if_exists="replace")
```

```python
# use the ids from the subredit data submissions to collect comments and write to bigquery.
comment_data = di.get_comments(reddit, current_data.id)
print("Comment data collected")

pandas_gbq.to_gbq(comment_data, table_com, project_id=project_id, if_exists="replace")
```
